```python
import pandas as pd
df = pd.read_csv("diamonds.csv")
```

```python
# Write a Pandas program to read the csv file and print the top 5 rows.
df[0:5]
```

```python
# Write a Pandas program to select a series from diamonds Dataset (Read in the form of DataFrame). Print the content of the series.
cut = df["cut"]
cut = pd.DataFrame(cut)
cut
```

```python
# Write a Pandas program to rename x column to length and y to width.
df = df.rename(columns={"x" : "length", "y" : "width"})
df
```

```python
# Write a Pandas program to sort the 'cut' column in ascending order and return it in a Series
df["cut"].sort_values(axis=0, ascending=True)
```

```python
# Write a Pandas program to find the diamonds that are with a Fair or Good or Premium.
result = df[df.cut.isin(['Fair', 'Good', 'Premium'])]
result
```

```python
# Write a Pandas program to calculate count, minimum, maximum price for each cut of diamonds DataFrame.
agg = df.groupby('cut').price.agg(['count', 'min', 'max'])
agg
```
